```python
import requests
import base64

params={
    'client_id': 'xxxxxxxxxxxxxxxxxxxxxxxxxxxxxx', 
    'client_secret':'xxxxxxxxxxxxxxxxxxxxxxxxx', #оба параметра можно найти в личном кабинете честного знака
    'user_id':'xxxxxxxxxxxxxxxxxxxxxxxxxxxx',  #отпечаток сертификата
    'auth_type':'SIGNED_CODE'
}

url="https://markirovka.crpt.ru/api/v3/true-api/auth/key"
response=requests.get(url, params=params).json()
uuid=response['uuid']
print(response)
```

```python
url="https://markirovka.crpt.ru/api/v3/true-api/auth/simpleSignIn"
data_with_signature="" #data подписанная прикрепленной эп, с убранными переносами строк
data={
    "uuid": uuid,
    "data": data_with_signature
}
response=requests.post(url, json=data).json()
print(response)
token=response['token']
```

```python

url="https://markirovka.crpt.ru/api/v3/true-api/lk/documents/create?pg=8"

file_path = "" #Путь к файлу, который необходимо отправить 

with open(file_path, "rb") as file: #Читаем файл в бинарном виде
    file_bytes = file.read() 

base64_file_bytes = base64.b64encode(file_bytes)  #Переводим файл в base64

with open(file_path + "." + "base64", "wb") as file: #записываем его
    file.write(base64_file_bytes)

with open(file_path + "." + "base64", 'r') as file: #читаем фай в base64
     document_base64 = file.read()

headers={
    'Authorization': 'Bearer '+ token
}

file_path_sign="" #Подготовленная открепленная ЭП в base64

with open(file_path_sign, 'r') as sig: 
    sign=sig.read()


data={
    "document_format":"MANUAL",
    "product_document": document_base64,
    "type":"LP_FTS_INTRODUCE",
    "signature": sign
    }

response_doc=requests.post(url, headers=headers, json=data)
```
